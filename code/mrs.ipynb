```python
import pandas as pd
import numpy as np
import string
```

```python
df = pd.read_csv("dataset/movie_dataset.csv")
df.head()
```

```python
df_selected = df.dropna()
```

```python
df_selected[['title','original_title']].sample(15)
```

```python
df_selected.drop(
    columns=['id','vote_average','vote_count','status','release_date','revenue','runtime',
             'budget','backdrop_path','homepage','imdb_id','original_language','original_title'
             ,'poster_path','production_companies','production_countries','spoken_languages'],
    inplace=True)
```

```python
df_selected
```

```python
df_selected['keywords'] = df_selected['keywords'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))
df_selected['genres'] = df_selected['genres'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))
df_selected['overview'] = df_selected['overview'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))
df_selected['tagline'] = df_selected['tagline'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))

df_selected['overview']
```

```python
df_selected['content'] = df_selected['overview']+ ' ' + df_selected['tagline'] + ' ' + df_selected['keywords'] + ' ' + df_selected['genres']
```

```python
df_selected
```

```python
df_selected['content'] = df_selected['content'].str.lower()
df_selected['content']
```

```python
df_final = df_selected.drop(columns=['genres','overview','keywords','tagline'])
df_final['title'] = df_final['title'].str.lower()
df_final
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# what is TFidfVectorizer?
# It gives higher importance to important words and lower importance to common words.

tfidf = TfidfVectorizer(stop_words='english')
matrix = tfidf.fit_transform(df_final['content'])

matrix
```

```python
df_final.shape
```

```python
matrix.shape
```

```python
adult = df_final['adult'].values.reshape(-1,1)
popularity = df_final['popularity'].values.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
popularity = minmax.fit_transform(popularity)
```

```python
from scipy.sparse import hstack

final_matrix = hstack([adult,popularity,matrix])
final_matrix
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors()
fit_model = model.fit(final_matrix)
```

```python
def recommend(name):

    name = name.lower()
    if name not in df_final['title'].values:
        print("Movie not found")
        return
    
    point = df_final.loc[df_final['title'] == name,['adult','popularity','content']]


    matrix_trf = tfidf.transform(point['content'])
    point_adult = point['adult'].values.reshape(-1,1)
    point_pop = point['popularity'].values.reshape(-1,1)

    point_pop = minmax.transform(point_pop)

    final_point = hstack([point_adult,point_pop,matrix_trf])

    _, indices = model.kneighbors(final_point, n_neighbors=5)

    for i in indices[0]:
        print(df_final.iloc[i]['title'])
```

```python
recommend('iron man')
```
